```python
from numpy import genfromtxt
import numpy as np
my_data=genfromtxt('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',delimiter=',',dtype='unicode')
```

```python
np.random.shuffle(my_data)
my_data
```

```python
training_data=my_data[0:90,0:4]

test_data=my_data[90:150,0:4]
```

```python
tr=np.array(training_data,dtype='float64')
te=np.array(test_data,dtype='float64')
```

```python
test=np.transpose(te)
train=tr.reshape(90,4,1)
#np.shape(test)
#train=np.reshape(90,4,1)
#np.shape(train)
```

```python
diff=train-test

dist=diff**2
sum_of=np.sum(dist,axis=1)
```

```python
final_result=np.argmin(sum_of,axis=0)
label_index=np.transpose(final_result)
train_label=my_data[0:90,4]
test_label=my_data[90:150,4]
label_list=[]
for index in final_result:
    label_list.append(train_label[index])
    


```

```python
counter=0
for num in range(0,60):
    if(test_label[num]==label_list[num]):
        counter=counter+1
    
print(counter)
accuracy=(counter/60)*100
print(accuracy)
```
